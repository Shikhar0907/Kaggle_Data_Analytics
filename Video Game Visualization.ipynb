Publishers each year

```python
source_publishers_each_year_df = source_df.groupby(['Year', 'Publisher']).agg({'Publisher': {'Published_count': 'count'}})
source_publishers_each_year_df.head(30)
```

Best Sales each year

```python
best_global_sales_each_year = source_df.groupby(['Year', 'Publisher']).agg({'Global_Sales': 'sum'})
best_global_sales_each_year.sort_values(by=['Year','Global_Sales'], ascending=False, inplace=True)
best_global_sales_each_year = best_global_sales_each_year.reset_index(level=['Year', 'Publisher'])
```

Global sales change increase or drop

```python
nintendo_gsales_df = best_global_sales_each_year[best_global_sales_each_year['Publisher'] == 'Nintendo'].sort_values(by=['Year']).head(50)
nintendo_gsales_df['Globa_sales_diff'] = nintendo_gsales_df.Global_Sales.diff()
nintendo_gsales_df.plot(x='Year', y='Globa_sales_diff', kind='bar', figsize=(10,10))
```

get each genres sales for each year based.

```python
genre_sales_df = source_df.groupby(['Year', 'Genre']).sum().reset_index()
genre_sales_df = pd.pivot_table(genre_sales_df, columns=['Year'], index=['Genre'], values=['Global_Sales'], aggfunc='first')
genre_sales_df.columns = genre_sales_df.columns.map(lambda x: '_'.join([str(i).split('.')[0] for i in x][::-1]))
genre_sales_df.head(20)
```

get change percentage

```python
# get years
years_df = source_df.groupby('Year').first().reset_index()
years_list = years_df['Year'].to_list()
years_list = [str(x).split('.')[0] for x in years_list]
years_list
percentage_diff_df = genre_sales_df

# get percentage difference in global sales of genres
for i in range(1, len(years_list)-2):
    prev_column = years_list[i-1] + '_Global_Sales'
    curr_column = years_list[i] + '_Global_Sales'
    percentage_diff_df[curr_column + '_diff'] = percentage_diff_df.apply(lambda x: (x[curr_column]/x[prev_column]) * 100, axis=1)      
            
percentage_diff_df.head(30)
```

PS4 vs XOne Global Sales Comparison

```python
ps4_vs_xbox_df = source_df[source_df['Platform'].isin(['PS4', 'XOne'])]
temp_df = ps4_vs_xbox_df.groupby(['Year', 'Platform']).sum().reset_index()
temp_df = pd.pivot_table(temp_df, columns=['Platform'], index=['Year'], values=['Global_Sales'], aggfunc='first')
temp_df.columns = temp_df.columns.map(lambda x: '_'.join([str(i).split('.')[0] for i in x][::-1]))
temp_df.plot.line()
```
